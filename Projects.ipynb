```python
import pandas as pd
import numpy as np

import simplejson as json
```

```python
# Load the dataset
data_folder = 'Prelim_results_for_students/2050/'
ssp1cc = 'SSP1_cc.csv'
data = pd.read_csv(data_folder+'SSP1_cc.csv')

```

```python
data.head()
```

```python
calories = data[['Calories', 'lat','lon']]
```

```python
calories.head()
```

```python
calories = calories.dropna()
```

```python
# Reduce Calories values
cal_series = calories['Calories'].apply(lambda x:int(np.log(x)*1000))
```

```python
calories['Calories'] = cal_series
```

```python
calories.describe()
```

```python
calories = calories.round(3)
```

```python
calories.to_csv('reduced_cal_lat_long.csv', index= False)
```

```python
reduced_calories = pd.DataFrame(columns=['Calories','lon', 'lat'])
```

```python
reduced_calories
```

```python
nb_sample = 10
```

```python
for i in range(0,len(calories)-nb_sample, nb_sample):
    temp_cal = (calories[i:i+nb_sample].mean())[0]
    temp_lat = (calories[i:i+nb_sample].mean())[1]
    temp_long = (calories[i:i+nb_sample].mean())[2]
    temp_df = pd.DataFrame([[temp_cal, temp_long,-temp_lat]], columns = ['Calories','lon', 'lat'])
    reduced_calories = reduced_calories.append(temp_df, ignore_index=True)
```

```python
reduced_calories.shape
```

```python
reduced_calories.to_json('test_10.json', orient='records')
```

```python
in_file = 'test_10.json'
out_file = 'geotest_10.geojson'


data = json.load(open(in_file))

geojson = {
    "type": "FeatureCollection",
    "features": [
    {
        "type": "Feature",
        "geometry" : {
            "type": "Point",
            "coordinates": [d["lon"], d["lat"]],
            },
        "properties" : d["Calories"],
     } for d in data]
}


output = open(out_file, 'w')
json.dump(geojson, output)
```
